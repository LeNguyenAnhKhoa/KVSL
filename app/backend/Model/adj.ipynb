```python
import pandas as pd
import os
import json
```

```python
df = pd.read_csv("./cleared.csv")

# Lấy danh sách tên file trong cột VIDEO
video_list = df["VIDEO"].astype(str).tolist()

# Thư mục chứa video
video_dir = "./Videos"

# Duyệt qua tất cả file trong thư mục Videos
for file in os.listdir(video_dir):
    if file.endswith(".mp4") and file not in video_list:
        file_path = os.path.join(video_dir, file)
        os.remove(file_path)
        print(f"Đã xóa: {file}")
```
