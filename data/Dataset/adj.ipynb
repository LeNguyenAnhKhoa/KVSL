```python
import pandas as pd
import cv2
import os

df1 = pd.read_csv("./label.csv")
df2 = pd.read_csv("./cleared.csv")
```

```python
df = df2.merge(df1[['LABEL', 'VIDEO']], on='LABEL', how='left')
df.head(3)
```

```python
# Lọc các dòng có VIDEO trống (NaN hoặc empty string)
df_empty_video = df[df['VIDEO'].isna() | (df['VIDEO'] == '')]

# Lưu ra file CSV
df_empty_video.to_csv('fix.csv', index=False)
```

```python
# Lấy danh sách video trong df
videos_in_df = set(df['VIDEO'].tolist())

# Lấy danh sách file .mp4 trong folder Videos
videos_in_folder = set([f for f in os.listdir('./Videos') if f.endswith('.mp4')])

# Kiểm tra video nào trong df không có trong folder
missing_videos = videos_in_df - videos_in_folder

if missing_videos:
    print(f"Thiếu {len(missing_videos)} video:")
    for video in sorted(missing_videos):
        print(f"  - {video}")
else:
    print("✓ Tất cả video trong df đều có trong folder /Videos")

# Thống kê
print(f"\nTổng video trong df: {len(videos_in_df)}")
print(f"Tổng video trong folder: {len(videos_in_folder)}")
```

```python
df = df[df['TOPIC'] == 'Đồ vật']
df['ID'] = range(len(df))
df.to_csv("./data.csv", index=False)
```
